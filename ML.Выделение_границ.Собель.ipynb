```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open(r"image.png")
plt.imshow(img)
```

```python
import numpy as np

img_gray = np.mean(img, axis=-1)
plt.imshow(img_gray, cmap='gray')
```

```python
sobel_matrix_1 = np.array([[-1, 0 , 1], [-2, 0, 2], [-1, 0, 1]])/4
sobel_matrix_2 = sobel_matrix_1.T
B = np.zeros((len(img_gray), len(img_gray[0])))

for x in range(len(img_gray)):
  for y in range(len(img_gray[0])):
    Gx = 0
    Gy = 0
    if (x!=0 and y!=0 and y!=len(img_gray[0])-1 and x!=len(img_gray)-1):
      for i in range(len(sobel_matrix_1)):
        for j in range(len(sobel_matrix_1[0])):
          Gx += sobel_matrix_2[i][j]*img_gray[x+i-1][y+j-1]
          Gy += sobel_matrix_1[i][j]*img_gray[x+i-1][y+j-1]
      B[x][y] = ((Gx ** 2 + Gy ** 2) ** (0.5)) / 2
```

```python
B1 = np.copy(B)
for x in range(len(img_gray)):
  for y in range(len(img_gray[0])):
    if (B[x][y] < 15):
        B1[x][y] = 0
    else: B1[x][y] = 255 
```

```python
plt.imshow(B1, cmap='gray')
```
